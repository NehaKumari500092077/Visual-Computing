#Task 2: Logical Operation on Images (3 Marks)
Perform a logical XOR operation on two grayscale images. Implement the XOR operation manually without using library functions. Verify the result using OpenCV. Use a_1_task_2_a.png and a_1_task_2_b.png as inputs.

```python
import numpy as np
from PIL import Image
import cv2 as cv
from google.colab.patches import cv2_imshow

# Method to calculate xor between two images
def xor_manually(path1, path2):

  # load the images and convert it to GreyScale
  img1 = Image.open(path1).convert('L')
  img2 = Image.open(path2).convert('L')

  # xor operation between two image array
  xor_data = np.bitwise_xor(np.array(img1), np.array(img2))

  # Convert numpy array to Image and return
  return xor_data

# Method to calculate xor between two images using opencv library
def xor_opencv_library(path1, path2):

  #load the images and convert it to GreyScale
  img1 = cv.imread(path1, cv.IMREAD_GRAYSCALE)
  img2 = cv.imread(path2, cv.IMREAD_GRAYSCALE)

  # return the result of xor operation between two images
  return cv.bitwise_xor(img1, img2)

# Method to verify if two images are equal or not
def calculate_mse(img1, img2):
  return np.mean((np.array(img1) - np.array(img2)) ** 2)

# path of the images
input1_path = "/content/a_1_task_2_a.png"
input2_path = "/content/a_1_task_2_b.png"

# perform XOR operation manually
xor_manually_output_image = xor_manually(input1_path, input2_path)

# perform XOR operation using opencv library
xor_opencv_library_output_image = xor_opencv_library(input1_path, input2_path)

# Show Image created by manual xor operation
print("Image created by manual xor operation\n")
cv2_imshow(xor_manually_output_image)
print("\n")
# Show Image created by opencv library xor operation
print("Image created by opencv library xor operation\n")
cv2_imshow(xor_opencv_library_output_image)
print("\n")

#verify using mean square error
mse_result = calculate_mse(xor_manually_output_image, xor_opencv_library_output_image)
if(mse_result < 1):
  print(f"Image created by XOR operation using opencv library and manuallly is Equal with mse as :: {mse_result}")
else:
  print(f"Image created by XOR operation using opencv library and manuallly isnot  Equal with mse as :: {mse_result}")
```
